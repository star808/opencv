## 3D图
To create 3D plots using Matplotlib, you can use the 'mplot3d' toolkit, which provides functions and classes specifically for 3D lotting
Here's a basic example of how to create a simple 3D plot:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Create a figure and a 3D axis 创建图形和3D轴
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate data 生成数据
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the surface 绘制曲面
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title 设置标签和标题
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

# Show the plot 
plt.show()
```

## 极轴图

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
theta = np.linspace(0, 2*np.pi, 100)
r = np.sin(3*theta)

# Create polar plot
plt.figure()
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)

# Set title
ax.set_title('Polar Plot')

# Show the plot
plt.show()
```

## 多重网格

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x + np.pi/2)

# Create a figure and a grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot data on the subplots
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sin(x)')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cos(x)')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Sin(x + pi/2)')

# Hide the empty subplot
axes[1, 1].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

##灰度图

```python
from pylab import *

def f(x,y): return(1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=10
x=np.linspace(-3,3,4*n)
y=np.linspace(-3,3,3*n)
x,y=np.meshgrid(x,y)
imshow(f(x,y))
```
